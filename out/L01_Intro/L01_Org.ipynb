# Организационные вопросы

## Проведение занятий

### Платформа
В период неблагоприятных эпидемиологических условий занятия будут проводиться дистанционно на платформе Discord. Вам необходимо зарегистрироваться и войти в виртуальный сервер по инвайт-ссылке: https://discord.gg/pQnW29fgHK . После перехода на сервер курса, вам необходимо в строгом порядке переименовать свой никнейм на реальные ФИО. После этого вам будет присвоена роль, и вы автоматически окажетесь в своей группе в Discord. Удостоверьтесь что была присвоена верная роль группы (название одной из ролей должно быть вашим номером группы, посмотреть это можно рядом с Вашим ФИО)

### Подготовка к занятию

И лекционный, и практический материал будет выдаваться аспирантам в формате ссылки на блокнот на платформе [Google Colab](https://colab.research.google.com) через чат в дискорде. Для работы Colab необходимо иметь аккаунт в Google!

### Формат занятий

Для того чтобы попасть на занятие, в нужное время зайти в комнату "лекция" Вашей учебной группы. После присоединения к голосовому каналу, преподаватель будет демонстрировать экран, к которому можно подключиться, нажав правой кнопкой на имя преподавателя и выбрав "Подключиться к стриму".

Занятие будет состоять из лекционной и практической частей, в соотношении примерно 1:2. После окончания лекционной части аспиранты приступают к выполнению практических заданий.

### Сдача заданий
Сдача выполненных заданий проходит в виртуальной комнате "сдача работ". Подключившись к ней, необходимо будет транслировать экран Вышего компьютера, поясняя свои решения и отвечая на вопросы преподавателя. 

Задания нужно успеть сдать до конца занятия. Если преподаватель не успевает посмотреть и проверить выполненное задание, то необходимо отправить текущее состояние блокнота на конец занятия с помощью гугл формы. (Скачав блокнот из Colab в формате .ipynb). Время отправки должно быть до конца занятия. Время фиксируется при отправке гугл формы. Частично выполненные задания тоже будут оценены.

Название файла-решения по шаблону:  'Lномер лекции_Фамилия.ipynb',  например:  'L01_Intro_Иванов.ipynb'

### Оценка заданий
Оценка за каждое занятие ставится на усмотрение преподавателя. Шкала оценивания от 0 до 100 баллов.

В конце модуля (7 занятий) составляется рейтинг аспирантов в группе в соответствии с суммой баллов за все занятия модуля. Первые 5 человек в рейтинге будут получать стипендию в следующем модуле. Последние 1-2 человека могут быть отчислены за неуспеваемость.

### Списывание

Списывать у других обучающихся в своей, в чужой группе - нельзя. Код пишется самостоятельно. Консультироваться с одногруппниками можно и нужно, когда вы находите ошибки друг у друга, обмениваетесь идеями, а не блоками кода.

Если преподаватель считает, что существенная часть кода была списана, то тот, у кого списали и тот, кто списал, получают незачет по заданию.

Если Вы используете блоки кода из интернета, то это разрешено. Однако нужно всегда указывать ссылку на источник, чтобы наличие одинакового кода у разных аспирантов не выглядело списыванием.

## Стипендия

По результатам **отборочного контеста** первого потока 15 лучшим аспирантам будет назначена стипендия на первый модуль в размере 20 тысяч рублей ежемесячно. Первая выплата в первой половине февраля 2021 года.

Во втором модуле (через 7 занятий) будут получать стипедндию 5 лучших аспирантов в каждой из групп.
