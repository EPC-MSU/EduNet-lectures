## SHAP<a class="anchor" style="autocontent" id="SHAP"/><br>
<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-47.png" width="600">

Этот способ широко используется везде, в том числе в нейросетях. Нейросети больше не является черным ящиком благодаря этому подходу. 

Суть способа в том, что мы пытаемся специфичным образом оценить вклад каждого признака.


<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-48.png" width="500">

Рассмотрим изображение выше. Возьмем 3 признака — "возраст", "пол" и "работа". Допустим, средняя зарплата — 50 000. Используем переменную “возраст” для конкретного человека. Например, ему 18 лет, он студент, подрабатывает, поэтому зарплата у него меньше (допустим, 40000). Если мы используем параметр “пол”, и речь идет о девушке, то, учитывая дискриминацию в обществе, она получает чуть меньше. Дальше мы можем узнать, что этот человек работает на хорошей работе, где средняя зарплата — 100000. 

Здесь мы обращали внимание только на один признак. Также можно обращать внимание на пару признаков. 
Наконец, обратим внимание на все три признака и на их основании предскажем зарплату в 83000. 

Что можно с этим сделать? 


<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-49.png" width="500">


Допустим, мы хотим оценить вклад возраста в предсказание. Выше выделены ребра, которое добавляют возраст в модель. Эти ребра можно взвесить. Как? 

<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-50.png" width="500">


Поступим следующим образом. У нас есть уровни. Первый уровень, когда у нас ноль признаков, второй уровень, когда у нас один признак и так далее. Веса связей между двумя соседними уровнями должны быть суммарно равны единице.

<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-49.png" width="500">

Дальше возьмем эти ребра. 

-10 × ⅓

-9 × ⅙

-15 × ⅙

-12 × ⅓

<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-51.png" width="500">

Сложим получившиеся значения. Получится, что средний вклад возраста в получившееся предсказание -11.33k.
Это значит, что для данного индивида возраст снижает зарплату на -11.33k по сравнению со случаем, если бы мы не учитывали возраст.

Такие же вычисления можно провести для остальных признаков (“пол”, “работа”). Соответственно, мы для каждого признака получим какую-то величину. Во-первых, эта величина действительно хорошо характеризует, насколько важен признак. Во-вторых, эту величину можно эффективно считать. 

Кажется, что эту величину считать долго. Ведь если у вас 3 признака, то нужно построить 7 моделей. С увеличением количества признаков экспоненциально растет число моделей, которые надо посчитать. 

Для многих задач существует точный способ, который позволяет дереву решений быстро получить нужный ответ, при этом не строя много деревьев. Для случайного леса аналогично. В случае с нейронными сетями это сложнее, но тоже возможно. 


## Работа с категориальными признаками<a class="anchor" style="autocontent" id="Работа-с-категориальными-признаками"/><br>
Поговорим о том, как можно работать с категориальными признаками. Допустим, у нас есть некоторые категории. Например, в случае с биологией рассмотрим нуклеотиды — органические соединения (на изображении ниже они обозначены буквами A, G, T, C), которые кодируют определенные аминокислоты, отвечающие за то, какой будет цвет волос, цвет глаз и прочие, абсолютно разные признаки. Из этих букв возможны любые комбинации.

<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-72.png" width="450">

У нас есть информация о том, что в геноме в каком-то гене одна буква заменилась на другую. Буква А заменилась на G (далее —А/G), буква Т заменилась на C (далее —Т/С). Мы хотим предсказать, замена какого типа произошла и использовать это как признак. Как предоставить эту информацию модели? Большинство моделей (нейронные сети, линейные модели и прочие) не умеют работать с нечисловыми признаками (в данном случае с текстовыми данными). Необходимо перевести их в числа. 

В данном случае категории замен (А/G, T/C), мы хотим перевести в числа. Наихудший, но самый простой способ, который иногда работает — каждой замене сопоставить цифру. То есть замена А на G в геноме соответствует нулю, замена T на C единице и так далее.

Аналогичный пример: у нас есть красный, зеленый, фиолетовый цвета. Красный соответствует нулю, фиолетовый — единице, желтый — двум. 


В данном случае, когда мы кодируем замену таким образом, мы задаем неявную информацию о том, что одна замена больше другой. То есть получается, что A/G меньше, чем Т/C (в случае с цветом получается, что красный больше, чем зеленый, но меньше, чем фиолетовый). Зачастую это абсолютно бессмысленная информация. То есть, если человек выбирает цвет айфона, он не выбирает его по спектру. Нельзя сказать, что красный цвет больше или меньше желтого. Но вы ввели эту информацию в модель, и она ее будет как-то использовать, попытается сравнить несравнимое. 

Чтобы такой ситуации не возникало, можно сделать эту нумерацию следующим образом: возьмем все уровни признака, все возможные значения, которые он принимает. Для каждого из значений считаем, какая у нас в среднем предсказываемая величина. Сортируем по средним значениям предсказываемой величины, и только после этого в порядке сортировки присваиваем номера. Тогда это будет давать какой-то результат. Но у этого способа тоже есть минусы, рассмотрим вариации подхода подробнее чуть позже.


### One hot encoding<a class="anchor" style="autocontent" id="One-hot-encoding"/><br>
<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-73.png" width="450">


Этот способ применяется наиболее часто. Допустим, у нас есть три цвета: красный, желтый, зеленый. Вместо того, чтобы работать с одним текстовым признаком, создаем три признака: “Цвет является красным?”, “Цвет является желтым?”, “Цвет является зеленым?”. В итоге мы вместо одного нечислового признака получили три признака, которые могут принимать числовые значения (либо 1, либо 0). На вопрос, является ли цвет красным, можно ответить только “да” или “нет”. 

Метод называется “One hot encoding” потому, что мы из одного признака получаем три, при этом для них выполняются условия, что только один из этих признаков отличается от нуля, при этом равен единице. 

В целом подход хороший. Иногда его используют с небольшими модификациями.


### Справочная таблица (lookup-table)<a class="anchor" style="autocontent" id="Справочная-таблица-(lookup-table)"/><br>
<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-74.png" width="450">

Следующий подход — это немного усложненный Label Encoding (когда мы сопоставляли числа). Его отличие в том, что каждому признаку сопоставляется не одно число, а сразу много. Одно из названий способа — справочная таблица lookup table. 

Мы берем и каждому категориальному признаку сопоставляем набор значений, и дальше работаем уже с этим набором значений. Понятно, что у этого метода тоже есть минусы, подобные Label Encoding`у, но гораздо менее явные, потому что этот вектор чисел мы берем не случайно, а каким-то образом выводим. Например, этот вектор чисел может учить нейросеть. То есть она сама может выучивать как ей категориальный признак представить — тем или иным способом. 


### Mean Encoding<a class="anchor" style="autocontent" id="Mean-Encoding"/><br>
Данный способ заключается в том, что мы фактически берем и кодируем метки средним. Идея в следующем: кодировать каждую метку не категориальным признаком, каким-то числом, взятым из сортировки или еще из чего-то, а каждому значению признака напрямую сопоставить среднее значение предсказываемой величины.

<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-76.png" width="650">

Рассмотрим, что будет, если закодировать признак неудачным способом. На графике слева каждому городу сопоставлено какое-то число. Было 100 городов, им сопоставлены числа от 0 до 100. Задача: предсказать, вернет человек кредит или нет. Если посмотреть, как выглядит распределение для классов “Отдали кредит” и “Не отдали кредит”, видно, что существенной разницы между двумя классами в этой переменной нет. Значит, эта переменная будет абсолютно бессмысленной для модели, либо модели придется делать много дополнительных преобразований, чтобы как-то работать с этим признаком. 

Если же каждому городу сопоставить процент людей, которые отдали кредит, то разделение по этому признаку будет очень хорошим. Это видно на графике справа.

Но тут возникает много проблем. Рассмотрим одну из них. Допустим, у нас есть город. В обучающей выборке из этого города никто не отдал кредит. Город очень маленький, в нем всего 10 человек взяли кредит и никто не вернул. Такое вполне возможно. При этом в тестовой выборке ситуация обратная: в ней всего 2 человека, и они кредит отдали. Тогда модель вполне разумно выучит, что надо предсказывать, что в этом городе люди никогда не отдают кредиты. Для людей в тестовой выборке она предскажет, что они не вернут кредит, хотя они его вернули. Получается ошибка. 

Причем эту ошибку не надо исправлять тем, что вы разбиваете training test по-другому. Проблема глубже: если какое-то значение категориального признака в принципе редкое, и вы пытаетесь для этого значения оценить, какое число ему сопоставить, это число может стать идеальным признаком, то есть оно просто говорит в прямом виде, что у нас в y (игрек). Поэтому переменная не используется в хорошем смысле, то есть как способ дополнительно уточнить предсказание. Она используется для того, чтобы взять эту переменную и напрямую ее выдать на выходе. При этом все разумные переменные игнорируются.


<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-77.png" width="600">

Борются с этим по-разному. Можно бить вашу обучающую выборку на блоки: у вас есть блок, для которого вы пытаетесь сделать Mean Encoding, а Mean Encoding делите по другим блокам. Условно говоря, город Москва кодируется на основании того, какая доля людей в остальных блоках вернула кредит. Получается, что данный признак равен городу Москва, он вам об этом блоке напрямую ничего не говорит, потому что при его кодировании никак этот блок не использовался. Далее для каждого блока делается отдельное кодирование. Получается, что кодирование менее явное, и меньше вероятности, что возникнут идеальные признаки. Тут тоже бывают проблемы, в том числе с редкими признаками. 

<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-78.png" width="400">

Поэтому можно сгладить это представление, можно добавлять некое глобальное среднее, то есть заранее ухудшать признак. Результат немного улучшится. 

### Expanding mean<a class="anchor" style="autocontent" id="Expanding-mean"/><br>
Этот способ использует Catboost. У Catboost есть единственное достижение — он действительно очень хорошо работает с категориальным признаками. 

<img src ="http://edunet.kea.su/repo/src/L03_Classic_ML/img/L03_Classic_ML-79.png" width="400">


Как работает Expanding mean: объекты в обучающей выборке как-то упорядочены. Обычно они упорядочены случайным образом, но если нет, то перемешаем их. Дальше категорию объекта n будем шифровать с использованием информации о категориях всех объектов до него. При этом можем на каждом шаге мешать объекты заново и получать новое кодирование для каждой категории. Этот способ более устойчивый. 

Считается, что Catboost, за счет использования этого способа, действительно хорошо работает с категориальными признаками. Однако, как уже говорилось ранее, его почти не применяют потому, что большую часть времени он не выдает результаты, сильно отличные от других моделей, а если вам очень надо использовать категориальные признаки, можно это сделать вручную, и результат будет не сильно хуже.
