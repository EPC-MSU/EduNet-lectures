```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

https://github.com/NikolayOskolkov/HowUMAPWorks/blob/master/HowUMAPWorks.ipynb,

https://www.nature.com/articles/s41467-018-07582-3, 
Cancer Associated Fibroblasts (CAFs) scRNAseq data. - датасет с фибробластами разного типа 

```python
data = pd.read_table("CAFs.txt") 
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
from sklearn.manifold import TSNE
```

```python
X_red = TSNE(n_components=2).fit_transform(X)
plt.figure(figsize=(10,10))
plt.title("tSNE on Cancer Associated Fibroblasts (CAFs)")
sns.scatterplot(x=X_red[:, 0], y=X_red[:, 1], hue=y)
```

```python
import umap
```

```python
X_red = umap.UMAP(n_components=2).fit_transform(X)
```

```python
plt.figure(figsize=(10,10))

plt.title("UMAP on Cancer Associated Fibroblasts (CAFs)")
sns.scatterplot(x=X_red[:, 0], y=X_red[:, 1], hue=y)
```
